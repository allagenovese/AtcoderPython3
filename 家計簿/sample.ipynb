```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
%matplotlib inline 
```

```python
df = pd.read_csv('202011.csv')
df.head()
df['sum'] = 0
```

```python
num = 0
for index, row in df.iterrows():
    num += row['利用金額']
    print(num)
    df.at[index,'sum'] += num
```

```python
df.head()
```

```python
df.plot(x = '利用日', y = 'sum',)
plt.xlabel('date')
plt.ylabel('sum')
```

```python
df_new = df.groupby('項目').sum()[['利用金額']]
df_new.plot(kind='pie', x= '項目', y = '利用金額', counterclock = False, startangle=90, autopct="%1.1f%%", pctdistance=0.7,figsize=(20,6))
```
